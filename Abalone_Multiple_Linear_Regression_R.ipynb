```
library(ucimlr)
library(dplyr)
library(tidyverse)
# info on the ucimlr package https://ucimlr.netlify.app/

#################################################
#  the first dataset from UCIMLR is the data on abalones (marine snails)
abalone_df = data.frame(abalone)
summary(abalone_df)
str(abalone_df)
devtools::install_github("rsquaredacademy/olsrr")

pairs(abalone_df[,2:ncol(abalone_df)])
```

```
#Illogical Height and outlier height
abalone_df <- subset(abalone_df,!(abalone_df$height==0 | abalone_df$height==1.13 |abalone_df$height == 0.515))

#Checking weights
abalone_df <- subset(abalone_df,whole_weight>shucked_weight)
abalone_df <- subset(abalone_df,!(abalone_df$whole_weight < abalone_df$viscera_weight))
abalone_df <- subset(abalone_df,!(abalone_df$whole_weight < abalone_df$shell_weight))

#Checking if shell weight less than or equal to wet weight
abalone_df <- subset(abalone_df,!(abalone_df$whole_weight - abalone_df$shucked_weight <= abalone_df$shell_weight))

summary(abalone_df)
str(abalone_df)
plot(abalone_df)
```

```
cor(abalone_df[,2:9],method="pearson")
```

```
abalone_MLR = lm(rings ~ sex+length+diameter+height+whole_weight+shucked_weight+viscera_weight+shell_weight, data = abalone_df)
summary(abalone_MLR)
sum(is.na(abalone_df))
```

```
# find the best regression model using stepwise. add
#   variables with a p value less than .1.  show the final regression
#   model with the r output

olsrr::ols_step_both_p(abalone_MLR, prem = .05, pent = .05, progress = TRUE)
#model suggest sex+diameter+height+shucked_weight+viscera_weight+shell_weight+whole_weight as predictors
```

```
#run model on suggested predictors
abalone_MLR = lm(rings ~ sex+diameter+height+shucked_weight+viscera_weight+shell_weight+whole_weight, data = abalone_df)
summary(abalone_MLR)

plot(abalone_df$rings,predict(abalone_MLR))
plot(abalone_MLR)

plot(predict(abalone_MLR),abalone_df$ring)
abline(lm(abalone_df$ring~predict(abalone_MLR)), col = "red", lwd = 3)
```

```
# find the best regression model using stepwise. add
#   variables with a p value less than .1.  show the final regression
#   model with the r output
#using log transformation
abalone_df$logring = log(abalone_df$rings)
abalone_MLR = lm(logring ~ sex+diameter+height+shucked_weight+viscera_weight+shell_weight+whole_weight+length, data = abalone_df)
summary(abalone_MLR)

olsrr::ols_step_both_p(abalone_MLR, prem = .001, pent = .001, progress = TRUE)
#model suggest sex+diameter+height+shucked_weight+viscera_weight+shell_weight+whole_weight+length as predictors
#transform response with log because there is right skewed
#abalone_MLR is the same as suggested variables selection
```

```
abalone.res <- resid(abalone_MLR)
plot(abalone_df$shell_weight,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. shell_weight")

plot(abalone_df$shucked_weight,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. shucked_weight")

plot(abalone_df$height,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. height")

plot(abalone_df$whole_weight,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. whole_weight")

plot(abalone_df$diameter,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. diameter")

plot(abalone_df$viscera_weight,abalone.res,ylab="Residuals",xlab="Actual",main="Plot of Residuals vs. viscera_weight")

boxplot(abalone.res~abalone_df$sex,main="Residuals - Sex")
```

```
abalone.pred=predict(abalone_MLR)

# Residuals vs Predicted Values
plot(abalone.pred,abalone.res,ylab="Residuals",xlab="Predicted Values",main="Plot of Residuals vs. Predicted Values")
abline(0,0)

# Actual vs Predicted
plot(predict(abalone_MLR),abalone_df$logring)
abline(lm(abalone_df$logring~predict(abalone_MLR)), col = "red", lwd = 3)
```
